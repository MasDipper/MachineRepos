# Импорт библиотек

```python
import pandas as pd
import numpy as np
import json
import os
```

```python
result_list = {'Заголовок': [], 'Название компании': [], 'Описание': [], 'Рейтинг': [], 'Категория': [], 'Дата': [], 'Текст': []}
brak = []
```

```python
json_files = os.listdir(r'C:\Users\kazyh\OneDrive\Рабочий стол\Data2')
json_files
```

```python
company_data = []
with open(r'C:\Users\kazyh\OneDrive\Рабочий стол\CompanyDate\company.json', 'r', encoding='UTF-8') as file:
    company_data = json.load(file)
```

```python
path = r'C:\Users\kazyh\OneDrive\Рабочий стол\Data2'
```

```python
for json_file in json_files:
    with open(f'{path}/{json_file}', 'r', encoding='UTF-8') as file:
        try:    
            data_json = json.load(file)
            for item in range(len(data_json['refs'])):
                if data_json['refs'][item] == None:
                    continue
                    
                if data_json['info'] != None:
                    result_list['Заголовок'].append(None)
                    result_list['Рейтинг'].append(data_json['info']['rate'])
                    result_list['Категория'].append(', '.join(data_json['info']['industries']))
                else:
                    [result_list[i].append(None) for i in ['Заголовок', 'Рейтинг', 'Категория']]
                    
                for company_item in company_data:
                    if company_item['company_name'] == json_file.replace('.json', ''):
                        result_list['Название компании'].append(company_item['company_name'])
                        result_list['Описание'].append(company_item['description'])
                date_pub = f'{data_json['refs'][item][1]['day']} {data_json['refs'][item][1]['month']} {data_json['refs'][item][1]['time']}'
                result_list['Дата'].append(date_pub)
                text_pub = ''.join(data_json['refs'][item][0]).replace('\n', ' ').replace('\t', ' ').replace('\xa0', ' ').replace('\u200e', ' ').replace('\r', ' ')
                result_list['Текст'].append(text_pub)
        except Exception as e:
            print(e)
            brak.append(json_file)
```

```python
df = pd.DataFrame(data=result_list)
```
